```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Aman", "Priya", "Karan"],
    "age": [21, 22, 20, 23, 21],
    "marks": [85, 75, 50, 92, 67]
})
df["grade"] = np.where(df["marks"] > 80, "A",
                np.where(df["marks"] > 60, "B", "C"))
print(df) 
```

```python
df = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Aman", "Priya", "Karan"],
    "age": [21, 22, 20, 23, 21],
    "salary": [85000, 7500, 50000, 92000, 6700]
})
df1 = df[df["salary"] > 50000]
print(df1)
```

```python
df = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Aman", "Priya", "Karan"],
    "age": [21, 22, 20, 23, 21],
    "salary": [85000, 7500, 50000, 92000, 6700]
})
df1 = df.sort_values(by = "age", ascending=False)
print(df1)
```

```python
df = pd.DataFrame({
    "product": ["Laptop", "Mobile", "Tablet", "Mobile", "Laptop"],
    "region": ["North", "South", "North", "East", "South"],
    "sales": [50000, 30000, 20000, 15000, 40000]
})
result = df.groupby("region")["sales"].sum()
print(result)
```

```python
df = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Aman", "Priya", "Karan","Amit","Laksh","sidhharth","vikas","Ganesh","Gayatri"],
    "age": [21, 22, 20, np.nan,23, np.nan,21,34,56,23,28],
    "salary": [85000, 7500, 50000, 92000, 6700, np.nan, 9900, np.nan, 23000, np.nan, 45000]
})
df["age"] = df["age"].fillna(df["age"].mean())
df["salary"] = df["salary"].fillna(df["salary"].mean())
print(df)
```

```python
df = pd.DataFrame({
    "name": ["Rahul", "Sneha", "Aman", "Priya", "Karan","Amit","Laksh","sidhharth","vikas","Ganesh","Gayatri"],
    "age": [21, 22, 20, np.nan,23, np.nan,21,34,56,23,28],
    "salary": [85000, 7500, 50000, 92000, 6700, np.nan, 9900, np.nan, 23000, np.nan, 45000]
})
df.to_csv("mydata.csv", index=False)
new_df = pd.read_csv("mydata.csv")
print("Original DataFrame:")
print(df)
print("\nDataFrame after reading from CSV:")
print(new_df)
```
