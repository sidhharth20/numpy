```python
import numpy as np
arr = np.array([3, 4.5, 2+3j, 7, 9.2, 5+0j])
complex_check = np.iscomplex(arr)
real_check = np.isreal(arr)
x = 10
y = np.array([1, 2, 3])
is_scalar_x = np.isscalar(x)
is_scalar_y = np.isscalar(y)
print("Array:", arr)
print("Complex number check:", complex_check)
print("Real number check:", real_check)
print("Is x scalar?:", is_scalar_x)
print("Is y scalar?:", is_scalar_y)
```

```python
arr1 = np.array([1.0, 2.001, 3.0, 4.0])
arr2 = np.array([1.0, 2.0001, 3.0, 3.9999])
tolerance = 0.01
result = np.abs(arr1 - arr2) <= tolerance
print("Element-wise comparison within tolerance:", result)
print("Are all elements equal within tolerance?:", np.all(result))
```

```python
arr1 = np.array([5, 10, 15, 20])
arr2 = np.array([3, 10, 18, 20])
greater = arr1 > arr2
greater_equal = arr1 >= arr2
less = arr1 < arr2
less_equal = arr1 <= arr2
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Greater (arr1 > arr2):", greater)
print("Greater or Equal (arr1 >= arr2):", greater_equal)
print("Less (arr1 < arr2):", less)
print("Less or Equal (arr1 <= arr2):", less_equal)
```
