```python
import numpy as np 
arry = np.array([[12,56,7,812],[2,4,6,8]])
print(f"The shape is :{arry.shape}")
print(f"The number of elements: {arry.size}")
print(f"The number of Dimentions are: {arry.ndim}")
print(f"The DataType is: {arry.dtype}")
print(f"Conversion to string: {arry.astype(str)}")
print(f"Conversion to integer: {arry.astype(int)} ")
print(arry + 3)
print(arry * 3)
print(arry - 11)
print(arry % 2)
print(arry // 2)
print(arry ** 4)
```
